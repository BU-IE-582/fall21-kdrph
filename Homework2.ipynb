```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.decomposition import PCA
from sklearn.manifold import MDS
```

```python
df = pd.read_csv (r'C:\Users\kpehl\Desktop\data.csv')
```

```python
dfa=df[df["class"]=='a']
dfb=df[df["class"]=='b']
```

```python
plt.scatter(dfa["X1"], dfa["X2"], c='red')
plt.scatter(dfb["X1"], dfb["X2"], c='blue')
plt.show()
```

```python
## Task 1 - a
X=df[['X1', 'X2']].to_numpy()
```

```python
pca = PCA()
pca.fit(X)
```

```python
print(pca.explained_variance_ratio_)
print(pca.singular_values_)
```

```python
t=np.transpose(pca.transform(X))[0]
```

```python
plt.scatter(t,np.zeros(198))
plt.show
```

```python
embedding = MDS(n_components=2)
```
