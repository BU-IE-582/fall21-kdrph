#Task 1 – Curse of dimensionality and effect of sample size

## Import libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

## For D from 1 to 15 dimensions, simulate 1000 random D-dimensional points, where the value in each dimension is uniformly randomly distributed between -1 and +1. 

```python
points = {}
for dimension in range(1,16):
   points[dimension] = np.random.rand(1000, dimension)*2 - 1
```

##Part a : Calculate the fraction of these points that are within distance 1 of the origin, giving an approximation of the volume of the unit hypersphere to the hypercube inscribing it.

```python
def distance_to_origin(point):
    return np.sqrt(np.sum(point**2))
```

```python
fractions = {}
for key,value in points.items():
    distances = np.apply_along_axis(distance_to_origin,1,value)
    fraction = np.sum(distances<=1)/len(distances)
    fractions[key] = fraction
```

```python
dimensions = list(fractions.keys())
fractions_ = list(fractions.values())
plt.plot(dimensions,fractions_)
plt.xlabel("Dimension")
plt.ylabel("Fraction")
plt.title("Fraction of Points in Hyper Cube")
plt.show()
```

##Part b: Use the value of this fraction at D = 2 and D = 3 to get estimates for the value of pi (Π) as you know the area (for D = 2) and volume (for D = 3) formulae for these cases. 

```python
#fractions_[1] -> D=2 ,  fractions_[2] -> D=3  and pi calculated from area and volume formulas for circle 2d pi*r*r square 4*r*r we can find fraction*4=pi in 2-D  .
#For 3d v= (4/3)* π *r^3 for sphere v= 8 *π * r^3 for cube then pi = 6*fraction
pi_value_2d = 4*fractions_[1]
pi_value_3d = 6*fractions_[2]
print("π is estimated from dimension 2 fraction as: "+ str(pi_value_2d))
print("π is estimated from dimension 3 fraction as: "+ str(pi_value_3d))
```

## Part c: Perform the calculations in part (b) with larger sample sizes. You can use the following set: {5000,10000, 25000, 50000, 100000}. Visualize the estimated Π for D = 2 and D = 3 cases. Comment on your results

```python
sizes = [500, 1000, 5000, 8000, 10000,25000, 35000, 40000,50000, 70000, 80000, 100000]
pi_values_size_3d = {}
pi_values_size_2d = {}
points_size = {}
for size in sizes:
  for dimension in range(2,4):
    points[dimension] = np.random.rand(size, dimension)*2 - 1
  distances_2d = np.apply_along_axis(distance_to_origin,1,points[2])
  fraction_2d = np.sum(distances_2d<=1)/len(distances_2d)
  distances_3d = np.apply_along_axis(distance_to_origin,1,points[3])
  fraction_3d = np.sum(distances_3d<=1)/len(distances_3d)
  pi_value_2d = 4*fraction_2d
  pi_value_3d = 6*fraction_3d
  pi_values_size_2d[size] = pi_value_2d
  pi_values_size_3d[size] = pi_value_3d
```

```python
print(pi_values_size_3d)
print(pi_values_size_2d)
```

```python
pi_values_2d = list(pi_values_size_2d.values())
pi_values_3d = list(pi_values_size_3d.values())
plt.plot(sizes,pi_values_2d, label="D==2",marker = "o")
plt.plot(sizes,pi_values_3d, label="D==3",marker = "o")
plt.xlabel("Sizes")
plt.ylabel("Pi Values")
plt.title("Fraction of Points in Hyper Cube")
plt.legend()
plt.show()
```

## Part d:  Repeat this simulation, sampling 1000 D-dimensional points from 1 to 15 dimensions, where the value in each dimension is uniformly randomly distributed between -1 and +1. For each value of D, generate an additional 100 test instances and calculate the distance to each test instance’s nearest neighbor. Plot the average distance from the test instances to their nearest neighbors as a function of D.

```python
# Create D-dimensional samples

points = {}
test_data = {}

for dimension in range(1,16):
    points[dimension] = np.random.rand(1000,dimension)*2 - 1

for dimension in range(1,16):
    test_data[dimension] = np.random.rand(100,dimension)*2 - 1
```

```python
# This function calculate the distance between two points of the same dimension
def dist_between(point1,point2):
    
    distance = 0
    for i in range(0,len(point1)):
        distance += (point1[i]-point2[i])**2
    return distance**0.5
```

```python
# Calculate the mean distances to the nearest neighbors
mean_distances = {}
for dimension in range(1,16):
  min_distances = []   
  for point1 in test_data[dimension]:
      distance = []  
      for point2 in points[dimension]:
          distance.append(dist_between(point1, point2))            
      min_distances.append(min(distance)) 
  mean_distances[dimension] = np.mean(min_distances)
```

```python
dimensions = list(mean_distances.keys())
distances = list(mean_distances.values())
plt.plot(dimensions, distances, marker="o")
plt.xlabel("Dimensions")
plt.ylabel("Avarage Distance to NN")
plt.title("Avarage Distance to NN")
plt.show()
```

##Task c: For each channel, take the average of the columns and plot the average as a line plot for each channel on a single plot. 

```python
red_channel = image_r[:,:,0]
green_channel = image_g[:,:,1]
blue_channel = image_b[:,:,2]
```

```python
red_mean_list = []
for columns in red_channel:
  red_mean_list.append(np.mean(columns, axis = 0))

green_mean_list = []
for columns in green_channel:
    green_mean_list.append(np.mean(columns, axis = 0))
    
blue_mean_list = []
for columns in blue_channel:
    blue_mean_list.append(np.mean(columns, axis = 0))
```

##Task d: For each channel, subtract one half of the image from the other half

```python
# For the blue part
blue_first_half = image_b[:,:256,:]
blue_second_half = image_b[:,256:,:]
blue_new_second_half = blue_second_half-blue_first_half
blue_new_second_half = blue_new_second_half.clip(0)
new_blue_image = np.concatenate((blue_first_half,blue_new_second_half),axis=1)
plt.imshow(new_blue_image);
```
